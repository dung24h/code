```python
#b9
import ast
if __name__=='__main__':
    s=input()
    t=input()
    d1=ast.literal_eval(s)
    d2=ast.literal_eval(t)
    for x in d2:
        if(x in d1):
            d1[x]+=d2[x]
        else:
            d1[x]=d2[x]
    d=dict(sorted(d1.items()))
    print(d)

    
```

```python
#B10
from functools import cmp_to_key

def cmp(a, b):
    if(a[0]<b[0]):
        return -1
    else:
        return 1

if __name__=='__main__':
    a=list(map(int, input().split()))
    k=int(input())
    mp={}
    res={}
    for x in a:
        tmp=k-x
        if(tmp in mp.keys()):
            t=(tmp, x)
            res[t]=1
        mp[x]=1
    a=[x for x in res.keys()]
    a.sort(key=cmp_to_key(cmp))
    print(a)
```

```python
#b12
from collections import Counter

if __name__=='__main__':
   a=list(map(int, input().split()))
   mp=dict(Counter(a))
   fre=0
   value=0
   for x in mp.keys():
    if(mp[x]>fre):
        value=x
        fre=mp[x]
    elif(x<value and mp[x]==fre):
        value=x
   print(value)


   

```
