Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần. Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy. mẫu hàm: def binary_search(arr, left, right, target):

```python
def binary_search(arr, target):
    left , right = 0, len(arr)-1
    while left <= right :
        mid = (left + right) //2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            right = mid -1
        else:
            left = mid + 1
arr=list(map(int, input().split()))
target=int(input())
print(binary_search(arr,target)) 
```

W7A3:
Viết hàm thực hiện thuật toán bubble sort, Selection Sort để sắp xếp danh sách theo thứ tự tăng dần. Sắp xếp nổi bọt (Bubble Sort); Sắp xếp chọn (Selection Sort)

```python
#Bubble sort:
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
#Selection sort
def selection_sort(arr):
    n = len(arr)
    for i in range(0, n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    
li = list(map(int, input().split()))
selection_sort(li)
print(li)
    
    
```

W7A7:
Cho một danh sách các số nguyên nums, hãy tìm độ dài của dãy con tăng nghiêm ngặt dài nhất (không nhất thiết phải liền kề).

```python
def day_tang(lst):
    n = len(lst)
    if n == 0:
        return 0
    dp = [1] * n 
    for i in range(n):
        for j in range(i):
            if lst[j] < lst[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)
lst = eval(input())
print(day_tang(lst))
```

Bạn được cung cấp: một danh sách các khoảng thời gian interval = [[l1, r1], [l2, r2], ...] và một danh sách các truy vấn queries = [q1, q2, ...] Với mỗi truy vấn q, hãy tìm độ dài nhỏ nhất của khoảng thời gian [li, ri] sao cho li ≤ q ≤ ri.
Nếu không tồn tại khoảng thời gian nào như vậy, hãy trả về -1 cho truy vấn đó. solve in < O(m.n), hint: binary search in the queries list

```python
def bin_se(x, pa):
    le = 0
    ri = len(pa) - 1
    
    ans = -1
    while le <= ri:
        mid = (le + ri) // 2
        if x <= pa[mid][1]:
            if x >= pa[mid][0]:
                ans = pa[mid][1] - pa[mid][0] + 1
                ri = mid - 1
            else: le = mid + 1
        else:
            le = mid + 1
    return ans


pa = eval(input())
pa.sort(key = lambda x: (x[1], x[1] - x[0] + 1))
li = eval(input())
for x in li:
    print(bin_se(x, pa), end = " ")
```

W7A10:
Cho một danh sách các số nguyên nums và một số nguyên m, hãy tìm tổng mảng con lớn nhất theo modulo m.
Tức là, với bất kỳ mảng con nums[i:j] nào, hãy tính sum(nums[i:j]) % m và trả về kết quả lớn nhất có thể.
Gợi ý:
Duy trì tổng tiền tố (prefix sum) mod m và sử dụng tìm kiếm nhị phân để tìm tổng tiền tố nhỏ nhất chỉ lớn hơn tổng hiện tại một chút (để tối đa hóa hiệu số bao quanh).

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def xuli(node, val, m):
    ans = 0
    while True:
        if val < node.val:
            ans = max(ans, m - (node.val - val))
            if node.left == None:
                node.left = Node(val)
                break
            else:
                node = node.left
        else:
            ans = max(ans, val)
            if node.right == None:
                node.right = Node(val)
                break
            else: 
                node = node.right
        
    return ans

li = list(map(int, input().split()))
m = int(input())
s = li[0] % m
root = Node(s)
ans = 0
for i in range(len(li)):
    s += li[i]
    s %= m
    ans = max(ans, xuli(root, s, m))
print(ans)
```

W7A13:
Dãy hài hòa là một dãy thỏa mãn điều kiện giá trị lớn nhất và giá trị nhỏ nhất của dãy khác nhau đúng bằng \( 1 \).
Viết hàm trả về độ dài của dãy hài hòa con dài nhất có thể có trong mảng giá trị nguyên \( arr \) có \( n \) phần tử
Ví dụ:
Input:
- 1 3 2 2 5 2 3 7
Output:
5 ( Dãy hài hòa con dài nhất là [3,2,2,2,3])

```python
li = list(map(int, input().split()))
ans = 1
mp = {}

for i in range(len(li) - 1, -1, -1):
    cnt = max(mp.get(li[i] - 1, 0) + mp.get(li[i], 0) + 1, mp.get(li[i] + 1, 0) + mp.get(li[i], 0) + 1)
    mp[li[i]] = mp.get(li[i], 0) + 1
    ans = max(ans, cnt)

print(ans)
```
