```
def electro(x):
	p1=50*1500 
	p2=50*2000
	if(x>=0) and (x<=50): return 1500*x
	elif(x>=51) and (x<=100): return p1+(x-50)*2000
	else: return p1+p2+(x-100)*3000
```
