```python
#W4A7
import math
def pt(n):
    x=int(math.sqrt(n))
    for i in range(2, x+1):
        while(n%i==0):
            res=n
            n//=i
    if(n!=1): res=n
    return res

n=int(input())
print(pt(n))

```

```python
#W4A9

import math

def kt(n):
    a=[0 for i in range(0, 10)]

    while(n>0):
        x=n%10
        if(a[x]==0): a[x]=1
        else: return 0
        n//=10
    return 1

n=int(input())
r=int(math.sqrt(n))
ok=0
for i in range(1, r+1):
    z=i*i
    if(kt(z)==1):
        ok=1
        print(z, end=' ')
if(not(ok)): print('no number')
```

```python
#W4A10


def collatz(n):
    step=1
    while(n!=1):
        if(n%2==0):
            n//=2
        else:
            n=3*n+1
        step+=1
    return step

n=int(input())
res=0
step=0
for i in range(1, n+1):
    tmp=collatz(i)
    if(tmp>step):
        res=i
        step=tmp

print(res, step)
```

```python
#W4A13
import math

def tong(x):
    sum=1
    r=int(math.sqrt(x))
    for i in range(2, r+1):
        if(x%i==0):
            sum+=i+x//i
    return sum
a, b=map(int, input().split())
if(tong(a)==b and tong(b)==a):
    print('true')
else: print('false')
```

```python
#W4A14

def gcd(a, b):
    if(b==0): return a
    return gcd(b, a%b);

def gcd2(a, b):
    r=None
    while(b!=0):
        r=a%b
        a=b
        b=r
    return a


m, n=map(int, input().split())
print(gcd2(m, n))
```

```python
#W4A23

def tong(n):
    return n*(n+1)/2
def bs(n):
    l=1
    r=10**9
    res=None
    while(l<=r):
        m=(l+r)//2
        if(tong(m)>=n):
            r=m-1
        else:
            res=m
            l=m+1
    return res

n=int(input())
print(bs(n))
```
